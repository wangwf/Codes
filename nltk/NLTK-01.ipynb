```python
sentence = '''NLTK is a wonderful tool for teaching, and working in, computational linguistics using Python, and an amazing library to play with natural language.
'''

tokens = nltk.word_tokenize(sentence)
tokens
```

```python
# POS tag
tagged = nltk.pos_tag(tokens)
tagged
```

```python
# Named entities
entities = nltk.chunk.ne_chunk(tagged)
#entities
```
